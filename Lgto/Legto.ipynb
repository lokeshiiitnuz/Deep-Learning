```python
import pandas as pd
import numpy as np
import random
```

```python
list1 = 'AAA-BBB-CCC-DDD-EEE-FFF-GGG-HHH-III-JJJ-KKK-LLL-MMM-NNN-OOO-PPP'.split('-')
```

```python
list1
```

```python
a = []
b = []
c = []
d = []
```

```python
for i in range(2,52):
    if i%3 == 0:
        a.append(np.nan)
    else:
        a.append( random.choice(list1))
    if i%4 == 0:
        b.append(np.nan)
    else:
        b.append( random.choice(list1))
    if i%5 == 0:
        c.append(np.nan)
    else:
        c.append( random.choice(list1))
    if i%6 == 0:
        d.append(np.nan)
    else:
        d.append( random.choice(list1))
```

```python
df = pd.DataFrame({'Diag_1':a,'Diag_2':b,'Diag_3':c,'Diag_4':d,})
```

```python
df['ID'] = np.arange(len(df))
```

```python
df = df[['ID', 'Diag_1', 'Diag_2', 'Diag_3', 'Diag_4',]]
```

```python
df
```

```python
dict1 = {}
for i in list1:
    dict1[i] = 0
```

```python
dict1
```

```python
df_temp = df[['Diag_1', 'Diag_2', 'Diag_3', 'Diag_4',]]
```

```python
dict1 = {}
for i in list1:
    dict1[i] = 0
for j in df_temp.iloc[0]:
    dict1[j] +=1
df_1 = pd.DataFrame.from_records([dict1])
```

```python
df_main = pd.DataFrame()
for k in range(len(df)):
    dict1 = {}
    for i in list1:
        dict1[i] = 0
    for j in df_temp.iloc[k]:
        if str(j) == 'nan':
            continue
        dict1[j] +=1
    df_1 = pd.DataFrame.from_records([dict1])
    df_main = pd.concat([df_main,df_1])
df_main
```

```python
df_main.iloc[0].corr(df_main.iloc[1])
```

```python
for i in range(len(df_main)):
    for j in range(len(df_main)):
        top_corr = 0
        if i ==j:
            continue
        temp_corr = df_main.iloc[i].corr(df_main.iloc[j])
        if temp_corr > top_corr:
            top_corr = temp_corr
            temp_j = j
    print(f'For ID - {i} High Correlation is with row {temp_j}. Correlation == {np.round((top_corr*100),2)}%')
```
