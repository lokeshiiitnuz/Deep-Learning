```python
import numpy as np
```

```python
import pandas as pd
```

```python
a = np.random.randint(1,100,size=(75))
b = np.random.randint(1,100,size=(75))
c = np.random.randint(1,100,size=(75))
```

```python
d = 3*a + 1.5*b -4*c +100
```

```python
df = pd.DataFrame({'A':a, 'B':b, 'C':c, 'D':d})
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x = df.drop(['D'], axis=1)
y = df['D']
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Lets try with NN

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

```python
x.shape
```

```python
model.add(tf.keras.layers.Dense(10, input_shape=(3,)))
```

```python
model.add(tf.keras.layers.Dense(32, activation='relu'))
```

```python
model.add(tf.keras.layers.Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = x
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
mea = tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error")
```

```python
model.compile(loss='mean_squared_error', metrics= [mea], optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=500, shuffle=True, validation_split=0.1, batch_size=10)
```
